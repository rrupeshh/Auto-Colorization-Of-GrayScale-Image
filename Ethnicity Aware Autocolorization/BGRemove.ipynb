```python
import os
from rembg import remove
from PIL import Image

# Define the root directory for your dataset and output directory
input_root_dir = 'datasetF'
output_root_dir = 'datasetBGRemoved'

# Walk through the directory tree
for subdir, dirs, files in os.walk(input_root_dir):
    for file in files:
        # Check if the file is an image
        if file.endswith(('.jpg', '.jpeg', '.png')):
            # Full path of the current image
            input_path = os.path.join(subdir, file)
            
            # Create corresponding output directory
            output_subdir = subdir.replace(input_root_dir, output_root_dir)
            if not os.path.exists(output_subdir):
                os.makedirs(output_subdir)
                
            # Output path for the image with background removed
            output_path = os.path.join(output_subdir, file)
            
            # Process the image
            input_image = Image.open(input_path)
            output_image = remove(input_image)
            
            # Convert RGBA to RGB if needed
            if output_image.mode == 'RGBA':
                output_image = output_image.convert('RGB')
            
            # Save the processed image
            output_image.save(output_path)
            print(f'Processed and saved: {output_path}')
```
